```python
#create Won collumn - treat the game based on environment and players status
np.random.seed(1999)
Win = np.random.randint(2,size=len(df))

df['Win'] = Win
isWin = (Win==0)
df.loc[isWin,'WRank'] , df.loc[isWin,'LRank'] = df.loc[isWin,'LRank'], df.loc[isWin,'WRank']
df.loc[isWin,'WPts'] , df.loc[isWin,'LPts'] = df.loc[isWin,'LPts'], df.loc[isWin,'WPts']
df.loc[isWin,'Player1'] , df.loc[isWin,'Player2'] = df.loc[isWin,'Player2'], df.loc[isWin,'Player1']
```

```python
#factorize
df['Location']=pd.factorize(df['Location'])[0]
df['Court']=pd.factorize(df['Court'])[0]
df['Surface']=pd.factorize(df['Surface'])[0]
df['Round']=pd.factorize(df['Round'])[0]**2
```

### Adding extra criterions
1. Number of wins in the current year up until the game

```python
# create column with number of wins for each players until the date of the game
def nwins_player_before_date(name,cur_date): # wins until game
    c=0
    for i in df.loc[df['Winner']==name].values:
        if cur_date > i[2]:
            c+=1
    return c**2
```

2. Number of wins in total before game/Number of games in total before game

```python
def raport_player_before_date(name,cur_date): # wins/total games until game in the same year
    c=1
    d=1
    #search = df.loc[cur_date.to_pydatetime().year==pd.DatetimeIndex(df['Date']).year]
    for i in df.loc[df['Winner']==name].values:
        if cur_date > i[2]:
            c+=1
    for i in df.loc[df['Loser']==name].values:
        if cur_date > i[2]:
            d+=1
    return c/(c+d)**2
```

Add columns to table

```python
def calc_nwins_by_date(my_df): # add 2 cols consisting of player wins per dat year until game
    my_df["P1dW"]=[nwins_player_before_date(name[0],name[1]) for name in my_df[["Player1","Date"]].to_numpy()]
    my_df["P2dW"]=[nwins_player_before_date(name[0],name[1]) for name in my_df[["Player2","Date"]].to_numpy()]
    my_df["P1dR"]=[raport_player_before_date(name[0],name[1]) for name in my_df[["Player1","Date"]].to_numpy()]
    my_df["P2dR"]=[raport_player_before_date(name[0],name[1]) for name in my_df[["Player2","Date"]].to_numpy()]
    return my_df
```

```python
df = calc_nwins_by_date(df)
```

#### Simulate game

```python
def simulate_game(entries): 
    # gets a df slice with 2 games
    # returns an entry containing the winners of both passed entries
    testing = entries.loc[:,selected_features2]
    prediction = clf.predict(testing.iloc[0:2])
    result = df_ao.loc[:,selected_features2]
    result = df_ao.iloc[:1] # get a line of the table to then change the contents
    
    game1 = entries.iloc[0:1]
    game2 = entries.iloc[1:2]
    
    result.loc[:,'Player1']=game1['Player1'].values[0] if prediction[0]==1 else game1['Player2'].values[0]
    result.loc[:,'WRank']=game1['WRank'].values[0] if prediction[0]==1 else game1['LRank'].values[0]
    result.loc[:,'WPts']=game1['WPts'].values[0] if prediction[0]==1 else game1['LPts'].values[0]
    result.loc[:,'P1dW']=game1['P1dW'].values[0] if prediction[0]==1 else game1['P2dW'].values[0]
    result.loc[:,'P1dR']=game1['P1dR'].values[0] if prediction[0]==1 else game1['P2dR'].values[0]
    result.loc[:,'MaxW']=game1['MaxW'].values[0] if prediction[0]==1 else game1['MaxL'].values[0]
    
    result['Player2']=game2['Player1'].values[0] if prediction[1]==1 else game2['Player2'].values[0]
    result.loc[:,'LRank']=game2['WRank'].values[0] if prediction[1]==1 else game2['LRank'].values[0]
    result.loc[:,'LPts']=game2['WPts'].values[0] if prediction[1]==1 else game2['LPts'].values[0]
    result.loc[:,'P2dW']=game2['P1dW'].values[0] if prediction[1]==1 else game2['P2dW'].values[0]
    result.loc[:,'P2dR']=game2['P1dR'].values[0] if prediction[1]==1 else game2['P2dR'].values[0]
    result.loc[:,'PSL']=game2['PSW'].values[0] if prediction[1]==1 else game2['PSL'].values[0]
    
    result.loc[:,'Winner']='-'
    result.loc[:,'Loser']='-'
    return result
```

Naive prediction of Australian Open 2019 without considering any distribution rules or other system requirements

The players were selected in order from the dataset and matched in pairs of 2. Then I applied the same rule on the dataset that resulted from the winners from the previous round.

```python
final_game=df_ao_r1
rounds=[]
r_index = 1 # round index
while len(final_game)>1:
    result_table = pd.DataFrame()
    for i in range(0,len(final_game),2):
        result_table=pd.concat([result_table,simulate_game(final_game.iloc[i:i+2])])
    final_game = result_table
    rounds.append(final_game)
    r_index+=1

final_game
```
